## Two options for subplot

```python
#  Subplot option 1
fig , ((ax1, ax2), (ax3, ax4)) = plt.subplots(
                                        nrows=2,
                                        ncols=2,
                                        figsize=(10,5)
)
# Plot to each different axis ax1 corresponds to the first plot
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.hist(np.random.randn(1000));
ax4.bar(list(nut_buter_prices.keys()) , list(nut_buter_prices.values()));
```

```python
# Subplots option 2
fig , ax = plt.subplots(
                    nrows=2,
                    ncols=2,
                    figsize=(10,5)
)

# Plot to each different index
# here we are plotting based on the index
ax[0,0].plot(x, x/2);
ax[0,1].scatter(np.random.random(10), np.random.random(10));
ax[1,0].bar(nut_buter_prices.keys(), nut_buter_prices.values());
ax[1,1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib OO methiod)
*When plotting something quickly use the pyplot method
* When plotting something more advanced, use the OO method

```python
heart_disease = pd.read_csv("heart-disease.csv")
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Here we are using pyplot method
over_50.plot(
            kind="scatter",
            x="age",
            y="chol",
            c="target" # here c means color but it will depend on the column 'target'
);
```

```python
# Let use OO method mixed with pyplot
fig , ax = plt.subplots(figsize=(10,6))
over_50.plot(
    kind="scatter",
    x="age",
    y="chol",
    c="target",
    ax=ax
);

ax.set_xlim([45,100])
# In this case, by setting the x-axis limits from 45 to 100, you're restricting the range of ages that will be displayed on the plot.
# Only ages between 45 and 100 will be shown on the x-axis, ensuring that the scatter plot focuses on a specific age range.
```

```python
# OO method from scratch
fix , ax = plt.subplots(figsize=(10,6))
# Plot the data
scatter = ax.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"] # here the color is based on target 
)


# Customize the plot

ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
#ax.legend(*scatter.legend_elements(), title="Target");
```

```python
# create a subplot with  age , chol , thalach

fig, (ax0 , ax1) = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10,10)
)

# add data to ax0
scatter = ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"] # here the color is based on target 
)

# Customize the plot

ax0.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");
# We can differentiate the two graphs

# Add a horizontl line
ax0.axhline(over_50["chol"].mean(),
           linestyle='--'
           
          );


# add data to ax1
scatter = ax1.scatter(
    x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"] # here the color is based on target 
)

# Customize the plot

ax1.set(title="Heart Disease and Max Heart Levels",
      xlabel="Age",
      ylabel="Max Heart Levels");

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");
# We can differentiate the two graphs

# Add a horizontl line
ax1.axhline(over_50["thalach"].mean(),
           linestyle='--'
           
          );
```

```python
# create a subplot with  age , chol , thalach

# here we are gonna to customize the plot as they are using the same x 

fig, (ax0 , ax1) = plt.subplots(
    nrows=2,
    ncols=1,
    figsize=(10,10),
    sharex=True # here we are allowing them to share the same x
)

# add data to ax0
scatter = ax0.scatter(
    x=over_50["age"],
    y=over_50["chol"],
    c=over_50["target"] # here the color is based on target 
)

# Customize the plot

ax0.set(title="Heart Disease and Cholesterol Levels",
      # xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target");
# We can differentiate the two graphs

# Add a horizontl line
ax0.axhline(over_50["chol"].mean(),
           linestyle='--'
           
          );


# add data to ax1
scatter = ax1.scatter(
    x=over_50["age"],
    y=over_50["thalach"],
    c=over_50["target"] # here the color is based on target 
)

# Customize the plot

ax1.set(title="Heart Disease and Max Heart Levels",
      xlabel="Age",
      ylabel="Max Heart Levels");

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target");
# We can differentiate the two graphs

# Add a horizontl line
ax1.axhline(over_50["thalach"].mean(),
           linestyle='--'
           
          );

fig.suptitle("Heart Disease" ,fontsize=16, fontweight="bold"); # we are customizing the figure
```
