```python
import random
import math

teams = []
playing_teams = {'myself': None, 'enemy': None}

class Team:
    def __init__(self, name, attack, defense):
        self.name = name
        self.attack = attack
        self.defense = defense
        self.total_score = 0

    def info(self):
        print(f"{self.name}: Attack: {self.attack} / Defense: {self.defense}")

    def get_hit_rate(self):
        return random.randint(10, self.attack)

    def get_out_rate(self):
        return random.randint(10, self.defense)

def create_teams():
    global teams
    teams = [
        Team('Attackers', 80, 20),
        Team('Defenders', 30, 70),
        Team('Averages', 50, 50)
    ]

def show_teams():
    print('All teams info:')
    for index, team in enumerate(teams, start=1):
        print(f"{index}.", end=" ")
        team.info()

def choice_team(player):
    player_name = 'Your team' if player == 'myself' else 'Opponent\'s'
    choice_team_number = int(input(f"{player_name}, please select a team (1–3): ")) - 1
    playing_teams[player] = teams[choice_team_number]
    print(f"{player_name} is the team \"{playing_teams[player].name}\"")

def get_play_inning(inning):
    hit_rate = playing_teams['myself'].get_hit_rate() if inning == 'front' else playing_teams['enemy'].get_hit_rate()
    out_rate = playing_teams['enemy'].get_out_rate() if inning == 'front' else playing_teams['myself'].get_out_rate()
    inning_score = max((hit_rate - out_rate) // 10, 0)
    return inning_score

def play():
    create_teams()
    show_teams()
    choice_team('myself')
    choice_team('enemy')
    
    score_boards = [['_________|'], ['You     ｜'], ['Opponent|']]
    
    for i in range(9):
        # Update inning numbers
        for row in score_boards:
            row.append(f"{i + 1}｜")
        
        # Top half of the inning
        inning_score = get_play_inning('front')
        print(f'Inning {i + 1} (front): {inning_score}')
        score_boards[1].append(f"{inning_score}｜")
        playing_teams['myself'].total_score += inning_score
        
        # Bottom half of the inning
        inning_score = get_play_inning('back')
        print(f'Inning {i + 1} (back): {inning_score}')
        score_boards[2].append(f"{inning_score}｜")
        playing_teams['enemy'].total_score += inning_score
    
    # Final scores
    score_boards[0].append('R｜')
    score_boards[1].append(f"{playing_teams['myself'].total_score}｜")
    score_boards[2].append(f"{playing_teams['enemy'].total_score}｜")
    
    for row in score_boards:
        print("".join(row))
    
    # Print final scores
    print("\nFinal Scores:")
    print(f"You: {playing_teams['myself'].total_score}")
    print(f"Opponent: {playing_teams['enemy'].total_score}")

play()
```
