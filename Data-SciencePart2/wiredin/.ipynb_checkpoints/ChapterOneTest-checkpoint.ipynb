```python
import random

coin = ['head', 'tail']
results = {'win': 'win', 'lose': 'lose'}

def get_my_bet():
    print('Input your bet')
    input_message = ''
    index = 0
    for face in coin:
        input_message += str(index) + ':' + face
        if index < 1:
            input_message += ', '
        index += 1
    return input(input_message + ' ') 

def start_message():
    print('Start \'coin toss\'')

def is_correct_input(string):
    if string.isdigit():
        number = int(string)
        if number >= 0 and number <= 1:  # Ensure number is within the range of options
            return True
    return False 

def get_coin_side():
    return random.randint(0, 1)

def get_side_name(index):
    return coin[index]

def view_side(my_guess, answer):
    print('My bet is ' + get_side_name(my_guess))
    print('Coin is ' + get_side_name(answer))

def get_result(my_guess, answer):
    if my_guess == answer:
        return 'win'
    else:
        return 'lose'

def view_result(result):
    print(results[result])

def play():
    start_message()
    my_guess = get_my_bet()
    while not is_correct_input(my_guess):
        my_guess = get_my_bet()

    my_guess = int(my_guess)
    answer = get_coin_side()

    view_side(my_guess, answer)
    result = get_result(my_guess, answer)
    view_result(result)

play()

```
